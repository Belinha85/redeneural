```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
X,y = make_classification(n_samples=100000,n_features=6,n_classes=2,random_state=42)
```

```python
df = pd.DataFrame(X)
df['target']=y
df
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
modelo = LogisticRegression()
modelo.fit(X_train,y_train)
```

```python
acuracia = modelo.score(X_test,y_test)
acuracia
```

```python
# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state=42)

# Treinando o modelo de árvore de decisão
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Avaliando o desempenho do modelo no conjunto de teste
score = model.score(X_test, y_test)
print("Acurácia do modelo: {:.2f}%".format(score * 100))
```
